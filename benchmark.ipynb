```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def read_result(path, name):
    df = pd.read_csv(path)
    df['just_type'] = name
    df['time'] = df['time'] / 1000.0
    return df

style = dict(capsize=0.04, errwidth=1.2, ci='sd')
```

# No. PEAF Nodes = 4,5,6 - EtaToAll (Sample Size = 100)

```python
df1 = read_result("./evaluation/evaluation_comparison/results_approx.txt", "approx_0_01")
df2 = read_result("./evaluation/evaluation_comparison/results_approx_exp.txt", "approx_exp_0_01")
df3 = read_result("./evaluation/evaluation_comparison/results_exact.txt", "exact")
df4 = read_result("./evaluation/evaluation_comparison/results_exact_exp.txt", "exact_exp")

df_complete = pd.concat([df1, df2, df3, df4])
df_complete.describe()
```

## No. PEAF Nodes vs. Time (in seconds)

```python
plt.figure(figsize=(16, 6))
sns.barplot(x="peafNodes", y="time",
             hue="just_type", 
             data=df_complete, **style)
```

## Without `approx_0_01`

```python
plt.figure(figsize=(16, 6))
sns.barplot(x="peafNodes", y="time",
             hue="just_type", 
             data=pd.concat([df2, df3, df4]), **style)
```

## No. PEAF Nodes vs. No. Iterations

```python
plt.figure(figsize=(16, 6))
sns.barplot(x="peafNodes", y="iteration",
             hue="just_type", 
             data=pd.concat([df1]), **style)
```

```python
plt.figure(figsize=(16, 6))
sns.barplot(x="peafNodes", y="iteration",
             hue="just_type", 
             data=pd.concat([df3]), **style)
```
