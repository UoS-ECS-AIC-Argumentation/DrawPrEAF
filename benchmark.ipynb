```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def read_result(path, name):
    df = pd.read_csv(path)
    df['just_type'] = name
    df['time'] = df['time'] / 1000.0
    return df

style = dict(capsize=0.04, errwidth=1.2, ci='sd')
```

# No. DAF Nodes = 3-20 - EtaToTree (Sample Size = 100)

```python
df_exact = read_result("./evaluation/evaluation_3_20_0_01/results_exact.txt", "exact")
df_con_exact = read_result("./evaluation/evaluation_3_20_0_01/results_con_exact.txt", "con_exact")
df_approx = read_result("./evaluation/evaluation_3_20_0_01/results_approx_0_01.txt", "approx_0_01")
df_con_approx = read_result("./evaluation/evaluation_3_20_0_01/results_con_approx_0_01.txt", "con_approx_0_01")


df_complete = pd.concat([df_exact, df_con_exact, df_approx, df_con_approx])
df_complete.describe()
```

## No. PEAF Nodes vs. Time (in seconds)

```python
plt.figure(figsize=(16, 6))
sns.barplot(x="peafNodes", y="time",
             hue="just_type", 
             data=df_complete[df_complete["peafNodes"] <= 14], **style)
```

```python
plt.figure(figsize=(16, 6))
sns.barplot(x="peafNodes", y="time",
             hue="just_type", 
             data=df_complete, **style)
```

## Exact vs Concurrent Exact

```python
plt.figure(figsize=(16, 6))
exacts = pd.concat([df_exact[df_exact["peafNodes"] <= 12], df_con_exact])
sns.barplot(x="peafNodes", y="time",
             hue="just_type", 
             data=exacts, **style)
```

## Approx vs Concurrent Approx

```python
plt.figure(figsize=(16, 6))
sns.barplot(x="peafNodes", y="time",
             hue="just_type", 
             data=pd.concat([df_approx, df_con_approx]), **style)
```

## No. PEAF Nodes vs. No. Iterations

```python
plt.figure(figsize=(16, 6))
sns.barplot(x="peafNodes", y="iteration",
             hue="just_type", 
             data=pd.concat([df_exact, df_approx[df_approx["peafNodes"] <= 14]]), **style)
```

```python
plt.figure(figsize=(16, 6))
sns.barplot(x="peafNodes", y="iteration",
             hue="just_type", 
             data=pd.concat([df_approx, df_con_approx]), **style)
```
